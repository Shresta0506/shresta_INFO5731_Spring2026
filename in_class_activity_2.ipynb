```python
# Starter cell (optional)
import requests
import pandas as pd

API_URL = "https://remoteok.com/api"
```

```python
import requests
import pandas as pd

API_URL = "https://remoteok.com/api"

# 1) Fetch data
resp = requests.get(API_URL, headers={"User-Agent": "Mozilla/5.0"})
resp.raise_for_status()
data = resp.json()

# 2) RemoteOK returns a list; first item is usually metadata (legal)
jobs = [
    item for item in data
    if isinstance(item, dict) and (item.get("position") or item.get("title")) and item.get("company")
]

# Q1) How many job postings?
q1_count = len(jobs)

# Build a DataFrame (helps with Q2/Q3 and Q4)
df_all = pd.DataFrame({
    "title": [j.get("position") or j.get("title") for j in jobs],
    "company": [j.get("company") for j in jobs],
    "location": [(j.get("location") or "Unknown") for j in jobs],
    "salary": [j.get("salary") if j.get("salary") not in ("", None) else None for j in jobs],
    "tags": [
        ", ".join(j.get("tags", [])) if isinstance(j.get("tags", []), list) else str(j.get("tags"))
        for j in jobs
    ],
})

# Q2) Top 3 most frequent companies
top3_companies = df_all["company"].value_counts().head(3)

# Q3) How many job titles contain "Data" (case-insensitive)?
q3_data_titles = df_all["title"].fillna("").str.contains("data", case=False, na=False).sum()

# Q4) First 10 job postings (exactly 10 rows) with required columns
df10 = df_all[["title", "company", "location", "salary", "tags"]].head(10)

# Print results clearly
print("Q1) Job postings available:", q1_count)
print("\nQ2) Top 3 companies:\n", top3_companies)
print("\nQ3) Titles containing 'Data':", q3_data_titles)

print("\nQ4) First 10 job postings DataFrame:")
df10.head(10)
```

Write a Python program that prompts the user to enter two numbers and perform a division operation. Handle exceptions for both zero division and invalid input (non-numeric input). Display appropriate error messages for each type of exception and ensure the program does not crash due to these errors.

```python
try:

    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))


    result = num1 / num2

    print("Result:", result)


except ZeroDivisionError:
    print("Error: You cannot divide by zero.")


except ValueError:
    print("Error: Please enter valid numeric values.")


except Exception as e:
    print("Unexpected error:", e)

else:
    print("Division successful.")

finally:
    print("Program execution completed.")

```

 Q2( 4 points)
Define a base class called 'Vehicle' with attributes make and model. Create a derived class Car that inherits from Vehicle and has an additional attribute 'num_doors'. Demonstrate an example of creating an instance of the 'Car' class and accessing its attributes.

```python

class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model


class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        # Call parent class constructor
        super().__init__(make, model)
        self.num_doors = num_doors



my_car = Car("Toyota", "Camry", 4)


print("Car Make:", my_car.make)
print("Car Model:", my_car.model)
print("Number of Doors:", my_car.num_doors)

```

Question 3 ( 2 points)
Create a program that accepts a list of numbers as input and outputs a new list containing only the even numbers.

```python

numbers = input("Enter numbers separated by commas: ")

num_list = [int(num.strip()) for num in numbers.split(",")]

even_numbers = [num for num in num_list if num % 2 == 0]


print("Even numbers list:", even_numbers)

```
