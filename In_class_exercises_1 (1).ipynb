## Question 2 ( 2 point)
Create a Python program that interactively takes two numerical inputs from the user and swaps their values without utilizing an extra variable.

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print(f"Original numbers: num1 = {num1}, num2 = {num2}")

# Swapping without an extra variable using tuple assignment
num1, num2 = num2, num1

print(f"Swapped numbers: num1 = {num1}, num2 = {num2}")
```

## Question 4 ( 2 point)
Design a Python function to check if a number is prime.


```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 2 prime? {is_prime(2)}")
```

## Question 5 ( 2 point)
Write a Python program to concatenate following dictionaries to create a new one.

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60},

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

# Concatenate dictionaries using dictionary unpacking
combined_dict = {**dic1, **dic2, **dic3}

print(combined_dict)
```
