## Q1 (1 point) — Answer below

```python
# Q1 — ANSWER CELL
FILE_PATH = "/content/support_messages.txt"

# TODO: load the TAB-separated file into df
df = pd.read_csv(FILE_PATH, sep="\t")

# TODO: print df.shape and df.head(3)
print("Shape of dataset:", df.shape)
# Print first 3 rows
print("\nFirst 3 rows:")
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
# Q2 — ANSWER CELL

import re

# word_count: number of words
df["word_count"] = df["message"].apply(lambda x: len(str(x).split()))

# char_count: number of characters
df["char_count"] = df["message"].apply(lambda x: len(str(x)))

# num_count: number of digits (0–9)
df["num_count"] = df["message"].str.count(r"\d")

# upper_word_count: number of ALL-CAPS words
df["upper_word_count"] = df["message"].apply(
    lambda x: sum(1 for word in str(x).split() if word.isupper())
)

# Display full DataFrame
print(df)
```

## Q3 (3 points) — Answer below

```python
import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words("english"))

def clean_text(text: str) -> str:
    text = str(text)

    # 1) Lowercase
    text = text.lower()

    # 2) Remove punctuation/symbols (keep letters, numbers, spaces)
    text = re.sub(r"[^a-z0-9\s]", "", text)

    # 3) Remove English stopwords
    words = text.split()
    words = [word for word in words if word not in STOPWORDS]

    # 4) Remove extra spaces
    text = " ".join(words)
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Create clean column
df["clean"] = df["message"].apply(clean_text)

# Print original and clean for id=1 and id=4
for i in [1, 4]:
    print(f"\nID = {i}")
    print("Original:", df.loc[df["id"] == i, "message"].values[0])
    print("Clean   :", df.loc[df["id"] == i, "clean"].values[0])
```

## Q4 (2 points) — Answer below

```python
import numpy as np

# order_id pattern (case-insensitive)
order_pattern = r"ORD-\d{4}"

# email pattern
email_pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"

# Extract first occurrence of order_id
df["order_id"] = df["message"].str.extract(
    f"({order_pattern})",
    flags=re.IGNORECASE
)

# Extract first occurrence of email
df["email"] = df["message"].str.extract(f"({email_pattern})")

# Display required columns
print(df[["id", "order_id", "email"]])
```

## Q5 (1 point) — Answer below

```python
# Q5 — ANSWER CELL

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

# Initialize TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the clean column
tfidf_matrix = vectorizer.fit_transform(df["clean"])

# Convert to DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

# Compute average TF-IDF score for each word
avg_tfidf = tfidf_df.mean(axis=0)

# Get top 5 keywords
top_5 = avg_tfidf.sort_values(ascending=False).head(5)

print("Top 5 Keywords by Average TF-IDF:")
print(top_5)
```
